```python
import matplotlib.pyplot as plt

# Data for the bar chart
games = ['NBA 2k25', 'Warhammer 40,000: Space Marine 2', 'Helldivers II', 'Call of Duty: Black Ops 6', 'Concord', 'Dragon Age: The Veilguard', 'Black Myth: Wukong', 'Star Wars Outlaws', 'Silent Hill 2', 'Skull and Bones', 
  'Suicide Squad: Kill the Justice League', 'S.T.A.L.K.E.R 2: Hearth of Chornobyl', 'Dragon Ball: Sparkling! Zero', 'Tekken 8', "Dragon's Dogma 2", 'Like a Dragon: Infinite Wealth', 'Metaphor: ReFantazio']
sales = [0.354, 3.565, 19.182, 15.552, 0.016, 0.946, 29.992, 0.028, 0.475, 0.035, 0.250, 1.133, 1.462, 0.880, 19.777, 0.434, 0.552]
sales_Trimmed = [0.354, 3.565, 15.002, 15.000, 0.016, 0.946, 15.003, 0.028, 0.475, 0.035, 0.250, 1.133, 1.462, 0.880, 15.001, 0.434, 0.552]
sortedData = sorted (zip(sales, games))
sales, games = zip(*sortedData)

# Create the horizontal bar chart
plt.barh(games, sales, color='skyblue')
for spine in plt.gca().spines.values(): spine.set_visible(False)

# Add titles and labels
plt.title('Stima vendite delle major releases del 2024')
plt.xlabel('MILIONI di copie vendute')
plt.ylabel('')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
